### 알고리즘 설명
주어진 리스트의 중간 값을 찾습니다.
중간 값과 리스트의 첫 번째 값 그리고 마지막 값을 비교하여 중간값을 피벗으로 선택합니다.
리스트를 피벗을 기준으로 두 그룹으로 나눕니다. 왼쪽 그룹은 피벗보다 작거나 같은 값으로 구성되고 오른쪽 그룹은 피벗보다 큰 값으로 구성됩니다. 이때 왼쪽 그룹의 끝 지점과 오른쪽 그룹의 시작 지점을 반환합니다.

### 코드 내용
파티션 함수에 리스트와 리스트의 처음 인덱스와 마지막 인덱스를 전달합니다.   
중간값을 기준으로 피벗을 선택합니다. 선택된 피벗 값은 출력됩니다.   
피벗을 기준으로 왼쪽과 오른쪽 그룹으로 나눕니다.   
피벗을 기준으로 왼쪽 그룹은 작거나 같은 값으로, 오른쪽 그룹은 큰 값으로 정렬됩니다.   
파티션 인덱스를 반환하고 이를 통해 리스트가 두 그룹으로 나누어진 위치를 확인할 수 있습니다.
